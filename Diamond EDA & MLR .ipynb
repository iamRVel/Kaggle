#Predictive modelling with scikit learn package

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_log_error,mean_squared_error, r2_score,mean_absolute_error
```

```python
sns.jointplot(x='volume', y='price', data= diam_df, size = 5)
```

#drop x,y,z variable since having multicollinearity 

```python
diam_df.drop(['x','y','z'], axis=1, inplace= True)
```

#Create dependent variable 

```python
XX = diam_df.drop(['price'], axis = 1)
YY = diam_df['price']
```

#Spliting the dataset as train and test.

```python
XX_train, XX_test, YY_train, YY_test = train_test_split(XX, YY, test_size = 0.3, random_state = 123)
```

#Apply feature scaling

```python
SC = StandardScaler()
XX_train = SC.fit_transform(XX_train)
XX_test = SC.transform(XX_test)
```

#Linear Regression

```python
clf_lr = LinearRegression()
clf_lr.fit(XX_train , YY_train)
accuracies = cross_val_score(estimator = clf_lr, X = XX_train, y = YY_train, cv = 5,verbose = 1)
YY_pred = clf_lr.predict(XX_test)
print('')
print('####### Linear Regression #######')
print('Score : %.4f' % clf_lr.score(XX_test, YY_test))
print(accuracies)

mse = mean_squared_error(YY_test, YY_pred)
mae = mean_absolute_error(YY_test, YY_pred)
rmse = mean_squared_error(YY_test, YY_pred)**0.5
r2 = r2_score(YY_test, YY_pred)

print('')
print('MSE    : %0.2f ' % mse)
print('MAE    : %0.2f ' % mae)
print('RMSE   : %0.2f ' % rmse)
print('R2     : %0.2f ' % r2)
```

#Random Forest regression

```python
clf_rf = RandomForestRegressor()
clf_rf.fit(XX_train , YY_train)
accuracies = cross_val_score(estimator = clf_rf, X = XX_train, y = YY_train, cv = 5,verbose = 1)
YYRF_pred = clf_rf.predict(XX_test)
print('')
print('###### Random Forest ######')
print('Score : %.4f' % clf_rf.score(XX_test, YY_test))
print(accuracies)

mse = mean_squared_error(YY_test, YYRF_pred)
mae = mean_absolute_error(YY_test, YYRF_pred)
rmse = mean_squared_error(YY_test, YYRF_pred)**0.5
r2 = r2_score(YY_test, YYRF_pred)

print('')
print('MSE    : %0.2f ' % mse)
print('MAE    : %0.2f ' % mae)
print('RMSE   : %0.2f ' % rmse)
print('R2     : %0.2f ' % r2)
```

From the given dataset I have performed explorative data analysis(EDA) and vizualising the plot with various packages to see the data pattern and

Following EDA I build a predictive modelling chosen only Linear and Random forest regression to be performed. From the model random forest regression modelling produces highest R2 score(98%).
